```python
import pandas as pd

# Replace 'your_file.csv' with the actual name of your CSV file
file_path = 'guardian_environment_news.csv'

# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFra
print(df.head())
```

```python
# prompt: drop column the first and last columns

df = df.drop(df.columns[[0, -1]], axis=1)
```

```python
df.head()
```

```python
# prompt: drop column with index 1

df = df.drop(df.columns[[1]], axis=1)
```

```python
df.head()
```

```python
# prompt: check if there are any null values

df.isnull().sum()
```

```python
# prompt: since there are only two columns both now. i want to merge these two columns respectively.

df['article'] = df[df.columns[0]].astype(str) + " " + df[df.columns[1]].astype(str)
df = df.drop(df.columns[[0, 1]], axis=1)
df.head()
```

```python
# prompt: count the nulls

print(df.isnull().sum())
```

```python
# prompt: print head
df.head()
```

```python
# prompt: print the number of columns and rows

print(df.shape)
```

```python
# prompt: print the average number of words per row

import numpy as np

# Calculate the number of words in each row
df['word_count'] = df['article'].apply(lambda x: len(x.split()))

# Calculate the average number of words per row
average_word_count = np.mean(df['word_count'])

# Print the average number of words per row
print(f"Average number of words per row: {average_word_count}")
```

```python
# prompt: make a list of of most used words (actual words not articles or prepositions)

import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
punctuations="?:!.,;’“”–()%"
stop_words = set(stopwords.words('english'))
words = []

for article in df['article']:
    tokenized_words = word_tokenize(article)
    for word in tokenized_words:
        if word.lower() not in stop_words and word.lower() not in punctuations:
            words.append(word.lower())

from collections import Counter

word_counts = Counter(words)
most_used_words = word_counts.most_common(10)

print(f"Most used words: {most_used_words}")
```
